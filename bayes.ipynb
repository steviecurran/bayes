## Baysian Test for Probability of Events

**E.g. testing postive for a disease, flipping a sequence of heads and tails...**

```python
import numpy as np
```

```python
print("Will calculate P(A|B) for multiple trials for percentages or decimals")

PA = float(input("\nP(A)? E.g. fraction of population with disease, probability of heads: "))
TP = float(input("P(B|A)? E.g. true positive rate: "))
FP = float(input("P(B|A')? e.g. false positive rate: "))
```

```python
if (PA >1) | (TP > 1) | (FP >1):
    TP = TP/100; FP = FP/100; PA = PA/100;
PAcomp = 1-PA
AB = (PA*TP)/(PA*TP + PAcomp*FP)
```

```python
i = 1 # TO COUNT
print("---------------------------------------------------------")
print("After %d test, P(A|B) =  %1.3f [%1.3f %%]" %(i, AB,AB*100))
print("---------------------------------------------------------")
```

```python
ans = str(input("Another test with the updated prior [y/n]? "))
while ans == 'y':
    i = i+1
    PA = AB
    o = str(input("Outcome positive (e.g. another heads) or negative (e.g. tails) [p/n]: "))
    if (o == "n") | (o == "N"): 
        TP = 1 - TP #P(T|C1)
        FP = 1 - FP #P(T|C1comp)

    PAcomp = 1 - PA #P(C1comp) = P(C2)
    AB = (PA*TP)/(PA*TP + PAcomp*FP)
    print("-------------------------------------------------------------------------")
    print("After %d tests, P(A|B) =  %1.3f [%1.3f %%]" %(i, AB,AB*100))
    print("-------------------------------------------------------------------------")
    ans = str(input("Another test with the updated prior [y/n]? "))
```
